```python
import numpy as np
import pandas as pd
from sklearn.model_selection import cross_val_score
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestClassifier
```

```python
df=pd.read_csv("/content/heart.csv")
```

```python
X=df.drop(columns="target",axis=1)
y=df["target"]
```

```python
models=[SVC(kernel="linear"),RandomForestClassifier(random_state=42),LogisticRegression(max_iter=1000),KNeighborsClassifier()]
```

```python
def Traning_model_crosscv():
  for model in models:
    cv_score=cross_val_score(model,X,y,cv=5,scoring="accuracy")
    cv_mean=cv_score.mean()
    cv_final=round(cv_mean*100,2)
    print(f"Accuracy of {model} is {cv_final}%")
```

```python
Traning_model_crosscv()
```

```python
models_list=[LogisticRegression(max_iter=10000),SVC(),RandomForestClassifier(random_state=42),KNeighborsClassifier()]
```

```python
model_dict={
    "Logisticregression":{
        "C":[1,5,10,20]
    },
    "svm":{
        "kernel":["linear","poly","rbf","sigmoid"],
        "C":[1,5,10,20]
    },
    "RandomForestClassifier":{
        "n_estimators":[10,20,50,100]
    },
    "KNN":{
        "n_neighbors":[3,5,10]
    }
}
```

```python
param_keys=list(model_dict.keys())
```

```python
def Best_param_model(list_models,param_dict):
  i=0
  res=[]
  for model in list_models:
    keys=param_keys[i]
    param=param_dict[keys]
    search_cv=GridSearchCV(model,param,cv=5,scoring="accuracy")
    search_cv.fit(X,y)
    res.append({"model":model,"best_parameter":search_cv.best_params_,"best_score":search_cv.best_score_})
    i+=1
  res=pd.DataFrame(res,columns=["model","best_parameter","best_score"])
  return res
```

```python
Best_param_model(models_list,model_dict)
```
